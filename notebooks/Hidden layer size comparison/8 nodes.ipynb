## Testing hidden node accuracy

Training several RBMs with increasing number of visible nodes. For each RBM we test how increasing the number of hidden nodes affects accuracy.

```python

#Initializing starting values 
visible_size = 8

n_hid_start = 16
n_hid = n_hid_start
hid_inc = 4
total_hid = 4

seed = 42

#Finding true ground state energy and displaying it
np.random.seed(seed)

H = RandomHamiltonian(n=visible_size)
eig,_ = np.linalg.eigh(H)
E_truth = np.min(eig)
#print(f"Energy truth: {E_truth}")

energy_list = []
time_list =[]
n_hid_list =[]
```

```python
for i in range(total_hid):
    np.random.seed(seed)

    rbm = RBM(visible_size=visible_size, hidden_size=n_hid, hamiltonian=H, walker_steps=0)
    
    energy_list.append(rbm.train(iterations=1000, lr=0.01))
    time_list.append(rbm.train.run_time)
                     
    n_hid_list.append(n_hid)
                     
    n_hid += hid_inc

# Print loop    
    labels = [f"Hidden = {i}" for i in n_hid_list]
```

```python
labels.append("True energy")
plt.figure(i)
# plt.ioff() # uncomment to stop plotting in notebook

for energy in energy_list:
    plt.plot(energy)
    plt.xlabel('Gradient steps')
    plt.ylabel('Energy')

plt.title(f"Training of RBM with {visible_size} visible nodes")
plt.axhline(y=E_truth, color='k', linestyle='--')
plt.legend(labels)
```

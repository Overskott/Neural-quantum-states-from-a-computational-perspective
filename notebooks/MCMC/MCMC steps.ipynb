```python
import numpy as np
import copy

from src.nqs import RBM, RandomHamiltonian, IsingHamiltonian, IsingHamiltonianReduced
import src.utils as utils
from config_parser import get_config_file

import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
visible_size = 10
hidden_size = visible_size
visible_size_list = range(2,15)


error_list = []
warm_up_list = []


for i in visible_size_list:
    error = []
    warm_up = []
    counter=1
    visible_size = i
    walker_steps = (i**3)*100
    print(f"visible size: {visible_size}")
    
    
    
    for w in range(0, 11):
        warm_up_steps = walker_steps//10
        
        H = IsingHamiltonian(n=visible_size)
        rbm = RBM(visible_size=visible_size, hidden_size=hidden_size, hamiltonian=H, walker_steps=walker_steps)


        mcmc, _ = np.histogram([utils.binary_array_to_int(state) for state in rbm.mcmc_dist(warm_up_steps)],
                 density=True, 
                 bins=range(2**visible_size+1))

        exact= rbm.probability_dist()
        exact = exact.reshape(-1)
        
        
        error.append(utils.prob_error(exact, mcmc))
        warm_up.append(visible_size)
        print(f"Probability error: {error[-1]}")
        counter += 1
    
    error_list.append(error)
    warm_up_list.append(warm_up)
    
    
```
